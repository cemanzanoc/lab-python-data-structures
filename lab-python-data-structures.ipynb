## Exercise: Managing Customer Orders

As part of a business venture, you are starting an online store that sells various products. To ensure smooth operations, you need to develop a program that manages customer orders and inventory.

1. Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".

6. Print the products in the `customer_orders` set.

   Store these statistics in a tuple called `order_status`.

8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
   ```

9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

10. Print the updated inventory, displaying the quantity of each product on separate lines.

Solve the exercise by implementing the steps using the Python concepts of lists, dictionaries, sets, and basic input/output operations. 

````python
#Step 1.Define a list called products with different fashion clothes
products=["t-shirt","mug","hat","book","keychain"]

#2.Create an empty dictionary called inventory
inventory={}

#3.Ask the user to input the quantity of each product available in the inventory.
# Use the product names from the list as key in the inventory and assign the respective quantity as values
for item in products:
    quantity=(int(input(f"Introduce the quantity of {item} available in the inventory:")))
    if quantity < 0:
        print(f"Error: la cantidad de {item} no puede ser negativa. Se asignarÃ¡ un valor de 0.")
        quantity = 0
    inventory[item]=quantity 
    print(f"There are {quantity} {item}")

#4.Create an empty set
customer_orders=set()

#5.Ask the user to input the names of three products that the customer wants
for i in range(3):
    order=input(f"Introduce your wanted item {i+1} for your order, remember that it should be only this items {products}").lower()
        
    if order in products:
        customer_orders.add(order)
    
    else :
        print(f"{order} is not a valid product.Please introduce an item from our products")
        print(f"{products}")
    

#6.Print your customer_order set
print("\nYour order is:", customer_orders)

#7.Calculate statistics
#7.1 Total products ordered
total_products_ordered=len(customer_orders)
total_available_products=len(products)
print(f"\nTotal products ordered: {total_products_ordered}")

#7.2 Percentage of products ordered
percentage_ordered=(total_products_ordered/total_available_products)*100
print(f"Percentage ordered compared to the total available products {percentage_ordered}%")

#Store statistics in a tuple
order_status=(total_products_ordered,percentage_ordered)

#8.Print the order statistics in a concrete format
print(f"\nOrder Statistics:\nTotal Products Ordered: {total_products_ordered}\nPercentage of Products Ordered: {percentage_ordered}%")

#9.Update the inventory
inventory["t-shirt"]-=1
inventory["mug"]-=1
inventory["hat"]-=1
inventory["book"]-=1
inventory["keychain"]-=1

#10.Print the Updated inventory
print("\nUpdated inventory:")
print(f"There are {products[0]} {inventory["t-shirt"]}")
print(f"There are {products[1]} {inventory["mug"]}")
print(f"There are {products[2]} {inventory["hat"]}")
print(f"There are {products[3]} {inventory["book"]}")
print(f"There are {products[4]} {inventory["keychain"]}")
````
